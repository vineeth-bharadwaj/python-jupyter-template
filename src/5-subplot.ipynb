### Exercise 5: Subplot Layout

```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data
x = np.arange(1, 6)
y = x ** 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
sizes = [40, 25, 20, 15]
labels = ['Python', 'Java', 'C++', 'JavaScript']
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff6666']
```

```python
try:
    # Check if lengths of x and y are the same for line and scatter plots
    if len(x) != len(y):
        raise ValueError("x and y must have the same length for line and scatter plots.")
    
    # Check if lengths of categories and values are the same for bar chart
    if len(categories) != len(values):
        raise ValueError("Categories and values must have the same length for the bar chart.")
    
    # Check if lengths of labels and sizes are the same for pie chart
    if len(labels) != len(sizes):
        raise ValueError("Labels and sizes must have the same length for the pie chart.")
    
    # Creating subplots
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))
    
    # Line chart (1st subplot)
    axs[0, 0].plot(x, y, color='blue', marker='o')
    axs[0, 0].set_title('Line Chart')
    axs[0, 0].set_xlabel('x values')
    axs[0, 0].set_ylabel('y values')
    axs[0, 0].grid(True)
    
    # Bar chart (2nd subplot)
    colors = ['red', 'green', 'blue', 'purple']
    axs[0, 1].bar(categories, values, color=colors)
    axs[0, 1].set_title('Bar Chart')
    axs[0, 1].set_xlabel('Categories')
    axs[0, 1].set_ylabel('Values')
    
    # Scatter plot (3rd subplot)
    axs[1, 0].scatter(x, y, color='orange')
    axs[1, 0].set_title('Scatter Plot')
    axs[1, 0].set_xlabel('x values')
    axs[1, 0].set_ylabel('y values')
    axs[1, 0].grid(True)
    
    # Pie chart (4th subplot)
    axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    axs[1, 1].set_title('Pie Chart')
    
    # Adjust layout
    plt.tight_layout()
    plt.show()

except ValueError as ve:
    print(f"ValueError: {ve}")

except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
