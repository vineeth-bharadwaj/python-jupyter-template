### Exercise 4: Pie Chart with Percentages

```python
# Importing necessary libraries
import matplotlib.pyplot as plt
```

```python
# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff6666']
explode = (0.1, 0, 0, 0)
```

```python
# Error handling for data validity
try:
    # Check if lengths of labels and sizes match
    if len(labels) != len(sizes):
        raise ValueError("Length of labels and sizes must be the same.")
    
    # Check if all sizes are non-negative numbers
    if any(size < 0 for size in sizes):
        raise ValueError("Sizes must be non-negative numbers.")
    
    # Creating the pie chart
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
    plt.title('Programming Language Popularity')
    plt.show()

except ValueError as ve:
    print(f"Input Error: {ve}")
```
