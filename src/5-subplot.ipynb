### Exercise 5: Subplot Layout

```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Data
x = np.arange(1, 6)
y = x ** 2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
sizes = [40, 25, 20, 15]
labels = ['Python', 'Java', 'C++', 'JavaScript']
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff6666']
```

```python
# Creating the subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Bar chart in the second subplot
axs[0, 1].bar(categories, values, color='orange')
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Line chart in the first subplot
axs[0, 0].plot(x, y, color='blue', marker='o')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('x values')
axs[0, 0].set_ylabel('y values')
axs[0, 0].grid(True)

# Scatter plot in the third subplot
axs[1, 0].scatter(x, y, color='green')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('x values')
axs[1, 0].set_ylabel('y values')
axs[1, 0].grid(True)

# Pie chart in the fourth subplot
axs[1, 1].pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart')

# Adjusting layout
plt.tight_layout()

# Display the plots
plt.show()
```
