### Exercise 1: Line Plot Customization

Import necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

Set the x and y values

```python
# Data
x = np.linspace(0, 10, 100)
y = x ** 2
```

Plot the y values against x with plot() function

```python
# Plot
plt.plot(x, y, label='y = x^2', color='blue', linestyle='-', marker='o')

# Customization
plt.title('Line Plot of y = x^2')
plt.xlabel('x values')
plt.ylabel('y values')
plt.grid(True)
plt.legend()
```
