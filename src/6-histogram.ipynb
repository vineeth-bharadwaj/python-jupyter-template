### Exercise 6: Histogram Customization

```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import math
```

```python
# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
```

```python
# Error handling for empty or invalid data
if not isinstance(data, list) or not all(isinstance(i, (int, float)) for i in data):
    raise ValueError("Input data must be a list of numbers.")
    
if len(data) == 0:
    raise ValueError("Input data cannot be empty.")

# Check for NaN or Infinite values
invalid_values = [i for i in data if math.isnan(i) or math.isinf(i)]
if invalid_values:
    raise ValueError(f"Input data contains NaN or Infinite values: {invalid_values}")

# Creating the histogram
plt.hist(data, bins=5, color='purple', edgecolor='black')
plt.title('Histogram of Data')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
