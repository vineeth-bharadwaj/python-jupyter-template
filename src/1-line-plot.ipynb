### Exercise 1: Line Plot Customization

Import necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

Set the x and y values

```python
# Data
x = np.linspace(0, 10, 100)
y = x ** 2
```

Plot the y values against x with plot() function

```python
try:
    # Error handling for data validity and completeness
    if not isinstance(x, (list, np.ndarray)) or not isinstance(y, (list, np.ndarray)):
        raise TypeError("x and y should be lists or numpy arrays.")
    if len(x) != len(y):
        raise ValueError("x and y must have the same length.")
    if len(x) == 0 or len(y) == 0:
        raise ValueError("x and y cannot be empty.")
    
    # Check if all elements are numeric
    non_numericx = [i for i in x if not isinstance(i, (int, float))]
    non_numericy = [i for i in y if not isinstance(i, (int, float))]
    if non_numericx or non_numericy:
        raise ValueError(f"Input data contains non-numeric values")
    
    # Check for NaN or Infinite values
    invalid_values = [i for i in x if math.isnan(i) or math.isinf(i)]
    if invalid_values:
        raise ValueError(f"Input data contains NaN or Infinite values: {invalid_values}")
    invalid_values = [i for i in y if math.isnan(i) or math.isinf(i)]
    if invalid_values:
        raise ValueError(f"Input data contains NaN or Infinite values: {invalid_values}")
        
    # Create the plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='y = x²', color='blue', linestyle='-')
    plt.title('Line Plot of y = x²')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.legend()
    plt.show()
except Exception as e:
    print(f"Error: {e}")
```
