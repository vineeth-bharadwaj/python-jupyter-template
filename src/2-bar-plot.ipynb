### Exercise 2: Bar Plot with Colors

Import matplotlib

```python
import matplotlib.pyplot as plt
```

Set the data to be plotted

```python
# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'purple']
```

Plot the data

```python
try:
    if len(categories) != len(values):
        raise ValueError("Length of categories and values must be the same.")
    if len(categories) == 0 or len(values) == 0:
        raise ValueError("Input must be non empty")
    non_numeric = [i for i in values if not isinstance(i, (int, float))]
    if non_numeric:
        raise ValueError(f"Input values contains non-numeric data")
    
    # Create a bar plot
    colors = ['red', 'green', 'blue', 'orange']
    plt.bar(categories, values, color=colors)
    plt.title('Bar Plot of Categories vs. Values')
    plt.xlabel('Categories')
    plt.ylabel('Values')
    plt.show()

except ValueError as ve:
    print(f"Error: {ve}")

except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
